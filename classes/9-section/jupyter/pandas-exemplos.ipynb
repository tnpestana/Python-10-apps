```python
# ler json da web
dataframe7=pandas.read_json("http://pythonhow.com/supermarkets.json")
dataframe7
```

```python
# manipular o index apresentado (set_index(index) não altera o indice guardado, apenas apresenta um diferente)
dataframe7.set_index("Address")
```

```python
# label based indexing
list(dataframe7.loc[:,"Country"])
```

```python
# position based indexing
dataframe7.iloc[1:4,1:4]
```

```python
# ix[] enables position/label based indexing
dataframe7.ix[3,"Name"]
```

```python
# omitir valores (o argumento 1 indica que vamos eliminar uma coluna, caso seja uma row é usado o 0)
dataframe7.drop(2,0)
```

```python
# dataframe.shape devolve um tuplo com o numero de rows e columns. shape[0] devolve o primeiro valor, numero de rows
# esta linha de codigo cria uma nova coluna "Continent" e associa às rows existentes o valor "North America"
dataframe7["Continent"]=dataframe7.shape[0]*["North America"]
dataframe7
```

```python
# o método .T troca as rows com as columns de um dataframe
dataframe7_t=dataframe7.T
dataframe7_t
```

```python
# é útil para criar uma nova row utilizando o método anterior que cria uma coluna 
dataframe7_t[6]=dataframe7_t.shape[0]*["New value"]
dataframe7=dataframe7_t.T
dataframe7
```

```python
# apagar alterações criadas anteriormente
dataframe7=dataframe7.drop("New address", 0)
dataframe7=dataframe7.drop(6,0)
dataframe7
```
