```python
import pandas
df=pandas.read_csv("supermarkets.csv")
df
```

```python
# agregar os valores das colunas mencionadas na coluna "Address"
df["Address"]=df["Address"]+", "+df["City"]+", "+df["State"]+", "+df["Country"]
df
```

```python
# criar uma nova coluna "Coordinates" e usar o metodo apply() para iterar todos os campos e guardar os resultados do nom.geocode
df["Coordinates"]=df["Address"].apply(nom.geocode)
df
```

```python
# lambda representa uma função inline
# neste caso é utilizada para recorrer ao valor guardado em "Coordinates" e extrair a sua latitude
# o if final indica que caso não haja um objeto de coordenadas guardado é guardado um objeto nulo de forma a prevenir erros
df["Latitude"]=df["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df["Longitude"]=df["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df
```
